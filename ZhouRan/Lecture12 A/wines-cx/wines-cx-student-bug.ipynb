## Dataset of Wine Quality

We download a dataset of Wine Quality, and we *learn it* with `calysto` (auto-encode it).
The dataset is from [http://archive.ics.uci.edu/ml/datasets.html](http://archive.ics.uci.edu/ml/datasets.html), it is about red wine quality, this data has 11 factors to influence the wine quality. Let's try to learn it with  `calysto`.

```python
import pandas as pd
import numpy as np
df=pd.read_csv('winequality-red.csv',header=None)
df.head(20)
```

For the wine quality experiment, we will create a 11-14-5 network:

* input layer of 11 units
* hidden layer of 14
* output layer of number of kinds of quality

```python
from calysto.ai.conx import *

wine_net = Network()
wine_net.addLayers(11, 14, 5) #input -11, hidden-14, output-5
wine_net
```

Set the quality level as output, choose 100 lines.

```python
result = df.loc[1:100,11].values.astype(float).tolist()
```

```python
result[:15]
```

Set other 11 factors as input.

```python
train = df.loc[1:100,:10].values.astype(float).tolist()
len(train[0])
```

We can propagate activation through the network (from input layer, through the hidden layer, to output layer):

```python
wine_net.propagate(input= [6.9, 0.605, 0.12, 10.7, 0.073, 40, 83, 0.9993, 3.45, 0.52, 9.4])
```

To see how an untrained network works on the wine problem, we can just propagate the input activations for each input pattern:

```python
for item in train:
    for pattern in [item]:
        print(pattern, wine_net.propagate(input=pattern))
```

Now, let's train a network to determine the weights to compute the wine quality dataset:

```python
results = []
for re_item in result:
    results.append([re_item])
len(train[0])
```

Add layers, but I encountered problems here.

```python
# provide training patterns (inputs and outputs)
wine_net = Network()
wine_net.addLayers(11, 14, 5)
wine_net.setInputs(train)
wine_net.setOutputs(results)

# set learning parameters
wine_net.setEpsilon(5)
wine_net.setTolerance(0.2)
wine_net.setReportRate(1)

# learn
wine_net.train()
```

```python
for item in train:
    for pattern in [item]:
        print(pattern, wine_net.propagate(input=pattern))
```
